```python
from bs4 import BeautifulSoup as bs
import requests
import re
```

```python
webpage = requests.get("https://coinmarketcap.com/").text
```

```python
doc = bs(webpage, "html.parser")
```

```python
# Crypto-coin details are stored in tbody tag
tbody = doc.tbody
```

```python
print(tbody.prettify())
```

```python
# Gives a list of all the tags inside tbody
trees =  tbody.contents
print(trees)
```

```python
# This should print Ethereum
trees[0].next_sibling
```

```python
# This should print BitCoin
trees[1].previous_sibling
```

```python
# As it should, it prints <tbody> as the parent tag
trees[0].parent
```

```python
# Will give you the tags inside of the table rows only
list(trees[0].children)
```

```python
# Returns tags inside of the children tags as well
list(trees[0].descendants)
```

```python
print(trees[0].contents)
```

```python
for tree in trees:
  for c in tree.contents:
    print(c)
    print()
```

```python
prices = {}

for tree in trees[:10]:
  name, price = tree.contents[2:4]
  name, price = name.p.string, price.a.string

  prices[name] = price

print(prices)
```
